```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('creditcard.csv')
print(df.shape)
df.head()
```

```python
X = df.drop(columns='Class', axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=2)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, average_precision_score
)

def evaluate_model(model, X_train, y_train, X_test, y_test, name="Model"):
    model.fit(X_train, y_train)
    # Some models (like SVC without probability=True) don’t have predict_proba
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
    elif hasattr(model, "decision_function"):
        # convert scores to [0,1] via min-max for AUPRC (approx)
        import numpy as np
        scores = model.decision_function(X_test)
        smin, smax = scores.min(), scores.max()
        y_proba = (scores - smin) / (smax - smin + 1e-8)
    else:
        y_proba = None

    y_pred = model.predict(X_test)

    print(f"\n=== {name} ===")
    print("Accuracy: {:.4f}".format(accuracy_score(y_test, y_pred)))
    print("Classification Report:\n", classification_report(y_test, y_pred, digits=4))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    if y_proba is not None:
        print("ROC-AUC: {:.4f}".format(roc_auc_score(y_test, y_proba)))
        print("Avg Precision (PR-AUC): {:.4f}".format(average_precision_score(y_test, y_proba)))
    else:
        print("No probability scores available for ROC/PR metrics.")

    return model
```

```python
# ==== Make best_features.json (run once) ====
from sklearn.feature_selection import mutual_info_classif
import pandas as pd, numpy as np, json

# If X_train is not a DataFrame, make it one with column names
if not isinstance(X_train, pd.DataFrame):
    # Kaggle dataset typical columns:
    cols = ["Time"] + [f"V{i}" for i in range(1,29)] + ["Amount"]
    X_train = pd.DataFrame(X_train, columns=cols)

# y_train to 1D Series
if isinstance(y_train, pd.DataFrame):
    y_train = y_train.iloc[:,0]
y_train = pd.Series(y_train)

# Mutual information scores & top-k features
mi = mutual_info_classif(X_train, y_train, random_state=42, discrete_features=False)
mi_s = pd.Series(mi, index=X_train.columns).sort_values(ascending=False)

top_k = 6                     # 6 best features (change to 5/8/10 if you want)
best_feats = mi_s.head(top_k).index.tolist()

# Medians for ALL features (to fill rest automatically)
medians = X_train.median(numeric_only=True).to_dict()

with open("best_features.json", "w") as f:
    json.dump({"best_features": best_feats, "medians": medians}, f, indent=2)

print("✅ best_features.json saved")
print("Top features:", best_feats)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("clf", KNeighborsClassifier(n_neighbors=25, weights="distance", n_jobs=-1))
])

knn = evaluate_model(knn, X_train, y_train, X_test, y_test, name="KNN (k=25, distance)")
```

```python
from sklearn.svm import SVC

svm_rbf = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("clf", SVC(kernel="rbf", C=1.0, gamma="scale", class_weight="balanced", probability=True))
])

svm_rbf = evaluate_model(svm_rbf, X_train, y_train, X_test, y_test, name="SVM (RBF, balanced)")
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=8,
    min_samples_leaf=20,
    class_weight="balanced",
    random_state=42
)

dt = evaluate_model(dt, X_train, y_train, X_test, y_test, name="Decision Tree (balanced)")
```

```python
import pandas as pd

models = {
    "KNN": Pipeline([("scaler", StandardScaler()),
                     ("clf", KNeighborsClassifier(n_neighbors=25, weights="distance", n_jobs=-1))]),
    "SVM-RBF": Pipeline([("scaler", StandardScaler()),
                         ("clf", SVC(kernel="rbf", C=1.0, gamma="scale", class_weight="balanced", probability=True))]),
    "DecisionTree": DecisionTreeClassifier(max_depth=8, min_samples_leaf=20, class_weight="balanced", random_state=42),
}

rows = []
for name, m in models.items():
    m.fit(X_train, y_train)
    y_pred = m.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    if hasattr(m, "predict_proba"):
        y_proba = m.predict_proba(X_test)[:,1]
    elif hasattr(m, "decision_function"):
        scores = m.decision_function(X_test)
        smin, smax = scores.min(), scores.max()
        y_proba = (scores - smin) / (smax - smin + 1e-8)
    else:
        y_proba = None

    roc = roc_auc_score(y_test, y_proba) if y_proba is not None else float("nan")
    pr  = average_precision_score(y_test, y_proba) if y_proba is not None else float("nan")

    rows.append({"model": name, "accuracy": acc, "roc_auc": roc, "pr_auc": pr})

pd.DataFrame(rows).sort_values("pr_auc", ascending=False)
```

```python
import joblib

best_model = models["KNN"]
best_model.fit(X_train, y_train)

# Save the trained model
joblib.dump(best_model, "fraud_knn_model.pkl")
print("✅ Model saved as fraud_knn_model.pkl")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, precision_recall_curve, roc_auc_score, average_precision_score
import matplotlib.pyplot as plt

y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=4))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
print("PR-AUC:", average_precision_score(y_test, y_proba))
```

```python
fpr, tpr, _ = roc_curve(y_test, y_proba)
prec, rec, _ = precision_recall_curve(y_test, y_proba)

plt.figure(figsize=(5,4))
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()

plt.figure(figsize=(5,4))
plt.plot(rec, prec)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curve")
plt.show()
```

```python
prec, rec, th = precision_recall_curve(y_test, y_proba)
f1s = 2 * prec * rec / (prec + rec + 1e-8)
best_idx = f1s.argmax()
best_thr = th[best_idx]
print("Best threshold:", best_thr)

y_pred_thr = (y_proba >= best_thr).astype(int)
print("New Confusion Matrix:\n", confusion_matrix(y_test, y_pred_thr))
print("New Classification Report:\n", classification_report(y_test, y_pred_thr, digits=4))
```

```python
import pandas as pd
import joblib, json
import numpy as np

model = joblib.load("fraud_knn_model.pkl")
with open("fraud_threshold.json") as f:
    meta = json.load(f)

FINAL_THRESHOLD = meta["threshold"]
FEATURES = meta["features"]

def predict_fraud(df_new: pd.DataFrame):
    # ensure column order & missing cols
    for c in FEATURES:
        if c not in df_new.columns:
            df_new[c] = 0
    df_new = df_new[FEATURES]  # order columns

    proba = model.predict_proba(df_new)[:,1]
    pred  = (proba >= FINAL_THRESHOLD).astype(int)
    return pred, proba

# EXAMPLE
sample = X_test.iloc[:5].copy()
pred, proba = predict_fraud(sample)
list(zip(pred, np.round(proba,4)))[:5]
```
